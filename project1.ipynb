```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
stocks_tickers=["RELIANCE.NS","TCS.NS","TATASTEEL.NS","INFY.NS","ITC.NS","HDFCBANK.NS","SBIN.NS","LT.NS","BHARTIARTL.NS","HCLTECH.NS","WIPRO.NS","ASIANPAINT.NS","M&M.NS","HINDUNILVR.NS","MARUTI.NS","SUNPHARMA.NS","AXISBANK.NS","BAJFINANCE.NS","ICICIBANK.NS","KOTAKBANK.NS"]
```

```python
stocks=yf.download(stocks_tickers,start='2022-01-01',end='2023-01-01')['Adj Close']
returns=stocks.pct_change().dropna()
returns
```

```python
corr_matrix=returns.corr()
corr_matrix
```

```python
listt=[]

for x in range (20):
    for y in range (x+1,20):
        for z in range (y+1,20):
            for a in range (z+1,20):
                for b in range (a+1,20):
                    st1=stocks_tickers[x]
                    st2=stocks_tickers[y]
                    st3=stocks_tickers[z]
                    st4=stocks_tickers[a]
                    st5=stocks_tickers[b]
                    mean_corr=corr_matrix[st1][st2]+corr_matrix[st1][st3]+corr_matrix[st1][st4]+corr_matrix[st1][st5]+corr_matrix[st2][st3]+corr_matrix[st2][st4]+corr_matrix[st2][st5]+corr_matrix[st3][st4]+corr_matrix[st3][st5]+corr_matrix[st4][st5]
                    listt.append((st1,st2,st3,st4,st5,mean_corr))
```

```python
listt_df=pd.DataFrame(listt, columns=['Stock1','Stock2','Stock3','Stock4','Stock5','Corr_sum'])
final=listt_df.sort_values(by='Corr_sum', ascending=True)
final[0:10]
```

```python
final_stocks=final.iloc[0]
print(final_stocks)
```
